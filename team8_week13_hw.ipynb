桃園市各區 事故位置類別 數量

目標loc_110_total
* x橫的標題放'發生市區鄉鎮名稱' '(事故位置子類別名稱)'.....(共1+19欄)[目前看到最多事故位置子類別有19種(中壢區) 其它區不到19種]
* y縱的標題放'(發生市區鄉鎮名稱)'
            中壢區(loc[0])
            八德區(loc[1])
            ...
            (共13個區)(loc[12])
            
arr: 即x橫的標題(以中壢區來看 為1+19個欄位)
a[0 for i in range(13)]: total[i].column(1)         即事故位置table(含事故位置名稱&group的數量) 的column(1) array(純的事故位置group數量)
a2[0 for i in range(13)]: np.append(loc[i], a[i])   即作為loc_110_total向下延伸 橫的array('中壢區' '(事故位置group數量)' '(事故位置group數量)'...'(事故位置group數量)')

So, 其實各區應該先印出全部的事故位置table(含事故位置名稱&group的數量) 即total[0 for i in range(13)]
這樣才能整個整合XD
=> 要對應正確的事故位置名稱 而不是直接向下延伸rows(橫的)

```python
# 各區域+事故位置類別
location_110 = table_110.select(['發生市區鄉鎮名稱','事故位置子類別名稱'])
location_110
```

```python
# 各區域事故數量
location_110_total = table_110.group('發生市區鄉鎮名稱')
location_110_total
```

```python
# <各區域><事故位置類別> 數量統計表
loc = [0 for i in range(13)]
total = [0 for i in range(13)]
a = [0 for i in range(13)]
a2 = [0 for i in range(13)]
pd = [0 for i in range(13)]

for i in range(13):
    loc[i] = location_110_total.column(0).item(i)
    total[i] = location_110.where('發生市區鄉鎮名稱',are.equal_to(loc[i])).group('事故位置子類別名稱')
    pd[i] = ((total[i].to_df()).rename(columns={'count':loc[i]})).set_index('事故位置子類別名稱')

for i in range(12):
    loc_110_total = pd[i].join(pd[i+1], how='outer')
    pd[i+1] = loc_110_total
loc_110_total = loc_110_total.astype('Int64') # float轉int
loc_110_total
loc_110_total_T = loc_110_total.T # 轉置(x,y軸交換)
loc_110_total_T
```

```python
# 單一區域 橫直方圖
# total[0]: 中壢區"事故位置子類別"數量統計表
for i in range(13):
  total_bar = ((total[i].to_df()).rename(columns={'count':loc[i]})).set_index('事故位置子類別名稱')
  total_bar.plot.barh(title='110年桃園市'+loc[i]+'事故發生市區_位置類別名稱件數', figsize=(15,10))
  plt.show()
```
